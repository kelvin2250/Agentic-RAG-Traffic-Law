```python
import shutil
from pathlib import Path

# Danh sách id cần giữ lại (giống list bạn đã cung cấp)
keep_ids = {
    "170620", "172475", "173920", "185666", "174961", "173895", "173738",
    "172771", "172772", "165031", "174932", "174892", "174890", "174888",
    "174886", "175618", "175620", "169924", "163057", "146222", "173121",
    "173068", "173166", "151447", "166844", "166909", "179198", "179201",
    "111475", "186156", "171712", "144830", "125142", "122401"
}

# Đường dẫn thư mục nguồn (chứa file .md) và đích
source_dir = Path("data/processed")
dest_dir = Path("data/processed_v2")

# Tạo thư mục đích nếu chưa tồn tại
dest_dir.mkdir(parents=True, exist_ok=True)

# Đếm và copy
copied_count = 0
for file_path in source_dir.glob("*.md"):  # chỉ lấy file .md
    file_id = file_path.stem   # lấy id từ tên file (không có .md)
    if file_id in keep_ids:
        dest_path = dest_dir / file_path.name
        shutil.copy2(file_path, dest_path)
        print(f"Đã copy: {file_path.name}")
        copied_count += 1

print(f"Hoàn tất! Đã copy {copied_count} file .md vào {dest_dir}")
```

```python
# ===================== CELL XỬ LÝ PDF VỚI TESSERACT + LLAMAPARSE =====================
import os
import re
import tempfile
import time
from pathlib import Path
import logging

import fitz  # PyMuPDF
import pdfplumber
import pytesseract
from pdf2image import convert_from_path
from dotenv import load_dotenv
from llama_parse import LlamaParse

load_dotenv()
logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")

# ---------------------- CẤU HÌNH ----------------------
DRY_RUN = False  # Đặt True nếu muốn test không gọi API thật
LLAMA_API_KEY = os.getenv("LLAMA_CLOUD_API_KEY")
GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")  # không dùng trong pipeline này, nhưng giữ

# Đường dẫn poppler (nếu đã cài bằng conda)
POPPLER_PATH = None
conda_prefix = os.path.dirname(os.path.dirname(sys.executable))  # thủ công
possible_poppler = os.path.join(conda_prefix, "Library", "bin")
if os.path.exists(possible_poppler) and any(f.endswith('.exe') for f in os.listdir(possible_poppler)):
    POPPLER_PATH = possible_poppler
    logging.info(f"✅ Tìm thấy poppler tại: {POPPLER_PATH}")
else:
    logging.warning("⚠️ Không tìm thấy poppler. Vui lòng chạy: conda install -c conda-forge poppler")

# Đường dẫn Tesseract
TESSERACT_PATH = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
if os.path.exists(TESSERACT_PATH):
    pytesseract.pytesseract.tesseract_cmd = TESSERACT_PATH
    logging.info(f"✅ Tesseract OK: {TESSERACT_PATH}")
else:
    logging.error("❌ Không tìm thấy Tesseract. Hãy cài từ: https://github.com/UB-Mannheim/tesseract/wiki")
    raise RuntimeError("Thiếu Tesseract")

# Ngôn ngữ OCR (nên cài tiếng Việt, nếu không thì dùng eng)
LANG = 'vie' if 'vie' in pytesseract.get_languages(config='') else 'eng'
logging.info(f"📖 Ngôn ngữ OCR: {LANG}")

# ---------------------- HÀM KIỂM TRA BẢNG THẬT ----------------------
def check_real_table(page_detector) -> bool:
    """Dùng pdfplumber để kiểm tra trang có bảng thực sự hay không"""
    tables = page_detector.find_tables()
    if not tables:
        return False
    for table in tables:
        extracted = table.extract()
        if extracted and len(extracted) >= 2 and len(extracted[0]) >= 2:
            return True
    return False

# ---------------------- OCR MỘT TRANG BẰNG TESSERACT ----------------------
def ocr_page_with_tesseract(pdf_path: str, page_num: int, dpi: int = 200) -> str:
    """
    Chuyển một trang PDF thành ảnh và OCR bằng Tesseract.
    pdf_path: đường dẫn file PDF gốc
    page_num: số trang (bắt đầu từ 0)
    dpi: 200 là cân bằng tốc độ/chất lượng
    """
    images = convert_from_path(
        pdf_path, dpi=dpi, first_page=page_num+1, last_page=page_num+1,
        poppler_path=POPPLER_PATH
    )
    if not images:
        return ""
    text = pytesseract.image_to_string(images[0], lang=LANG)
    # Làm sạch: xóa dòng chỉ toàn số (số trang), chuẩn hóa xuống dòng
    lines = text.split('\n')
    filtered_lines = []
    for line in lines:
        stripped = line.strip()
        if stripped and not re.fullmatch(r'[\d\s\.]+', stripped):
            filtered_lines.append(line)
    text = '\n'.join(filtered_lines)
    text = re.sub(r'\n\s*\n', '\n\n', text.strip())
    return text

# ---------------------- LLAMAPARSE CHO MỘT TRANG (CÓ BẢNG) ----------------------
llama_parser = None
def get_llama_parser():
    global llama_parser
    if llama_parser is None and LLAMA_API_KEY:
        llama_parser = LlamaParse(
            api_key=LLAMA_API_KEY,
            result_type="markdown",
            use_vendor_multimodal_model=True,
            language="vi",
        )
    return llama_parser

def parse_page_with_llamaparse(pdf_path: str, doc_reader: fitz.Document, page_num: int) -> str:
    """Trích xuất một trang PDF (có bảng) bằng LlamaParse, trả về markdown"""
    parser = get_llama_parser()
    if parser is None:
        logging.warning(f"Không có LlamaParse key, fallback sang OCR cho trang {page_num+1}")
        return ocr_page_with_tesseract(pdf_path, page_num)
    
    # Tạo file PDF tạm chỉ gồm trang cần xử lý
    doc_single = fitz.open()
    doc_single.insert_pdf(doc_reader, from_page=page_num, to_page=page_num)
    with tempfile.NamedTemporaryFile(delete=False, suffix=".pdf") as tmp:
        temp_pdf_path = tmp.name
    try:
        doc_single.save(temp_pdf_path)
        parsed = parser.load_data(temp_pdf_path)
        if parsed and len(parsed) > 0:
            return parsed[0].text.strip()
        else:
            return ""
    except Exception as e:
        logging.error(f"LlamaParse lỗi ở trang {page_num+1}: {e}")
        return ocr_page_with_tesseract(pdf_path, page_num)  # fallback
    finally:
        doc_single.close()
        try:
            os.unlink(temp_pdf_path)
        except:
            pass

# ---------------------- XỬ LÝ TOÀN BỘ FILE PDF ----------------------
def process_pdf_with_tesseract_and_llamaparse(pdf_path: str) -> str:
    """
    Xử lý một file PDF:
      - Mỗi trang: kiểm tra có bảng thật không.
      - Nếu có bảng -> dùng LlamaParse (giữ cấu trúc bảng).
      - Nếu không -> OCR bằng Tesseract.
    """
    if not os.path.exists(pdf_path):
        raise FileNotFoundError(f"Không tìm thấy file: {pdf_path}")
    
    final_markdown = []
    llamaparse_call_count = 0
    total_pages = 0
    
    try:
        with pdfplumber.open(pdf_path) as pdf_detector, fitz.open(pdf_path) as doc_reader:
            total_pages = len(pdf_detector.pages)
            for page_num, page_detector in enumerate(pdf_detector.pages):
                has_table = check_real_table(page_detector)
                
                if has_table and LLAMA_API_KEY:
                    logging.info(f"📊 Trang {page_num+1} có bảng → dùng LlamaParse")
                    if not DRY_RUN:
                        page_md = parse_page_with_llamaparse(pdf_path, doc_reader, page_num)
                        llamaparse_call_count += 1
                    else:
                        logging.info("[DRY RUN] Bỏ qua LlamaParse")
                        page_md = f"<!-- DRY RUN: LlamaParse cho trang {page_num+1} (có bảng) -->"
                    final_markdown.append(page_md)
                else:
                    # Không có bảng hoặc không có LlamaParse key → dùng Tesseract
                    if has_table and not LLAMA_API_KEY:
                        logging.warning(f"⚠️ Trang {page_num+1} có bảng nhưng thiếu LlamaParse API key → dùng OCR thường")
                    else:
                        logging.info(f"📄 Trang {page_num+1} không bảng → OCR bằng Tesseract")
                    
                    if not DRY_RUN:
                        ocr_text = ocr_page_with_tesseract(pdf_path, page_num)
                    else:
                        ocr_text = f"[DRY RUN] OCR trang {page_num+1}"
                    final_markdown.append(ocr_text)
    
    except Exception as e:
        logging.error(f"Lỗi xử lý PDF {pdf_path}: {e}")
        return ""
    
    # Báo cáo chi phí
    print(f"\n📊 Báo cáo xử lý: {os.path.basename(pdf_path)}")
    print(f"   - Tổng số trang: {total_pages}")
    print(f"   - Số trang có bảng (dùng LlamaParse): {llamaparse_call_count}")
    print(f"   - Số trang OCR bằng Tesseract: {total_pages - llamaparse_call_count}")
    if llamaparse_call_count > 0:
        print(f"   💰 LlamaParse: {llamaparse_call_count} trang (miễn phí 1000 trang/ngày)")
    
    return "\n\n".join(final_markdown)

# ===================== SỬ DỤNG =====================
# Thay đường dẫn file PDF của bạn vào đây
input_pdf = r"data\raw\179201.pdf"   # Ví dụ
output_md = r"output_ocr_llama.md"

if os.path.exists(input_pdf):
    md_content = process_pdf_with_tesseract_and_llamaparse(input_pdf)
    with open(output_md, "w", encoding="utf-8") as f:
        f.write(md_content)
    print(f"\n✅ Đã lưu kết quả vào: {output_md}")
    print("\n===== Xem trước 1500 ký tự đầu =====")
    print(md_content[:1500] + ("..." if len(md_content) > 1500 else ""))
else:
    print(f"❌ Không thấy file: {input_pdf}")
```
